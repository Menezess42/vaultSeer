# Use Path from pathlib

- Now: Decide which data structure will keep files and directories in the correct order

- The best approach is to use a n-area tree here

- N-Area tree node

```python
class nArea_treeNode:
    def __init__(self, path: Path) -> None:
        self.name = path.name
        self.path = path
        self.is_dir = path.is_dir()
        self.is_file = path.is_file()
        if self.is_file:
            aux = self.name.split(".")
            self.file_type = aux[-1]
        self.children: list[nArea_treeNode] = []

    def add_child(self, child: "nArea_treeNode") -> None:
        self.children.append(child)
```

- N-Area feeding

```python
# Dir to visit stack
dirQueue = []
dirTree = nArea_treeNode(dir_path)
print(dirTree)
print(type(dirTree))
```

```python
def nArea_feeding(dirTree: nArea_treeNode)-> None:
    dirQueue.append(dirTree)
    while dirQueue != []:
        cRoot = dirQueue.pop(0)
        print(cRoot)
        for name in cRoot.path.iterdir():
            newNode = nArea_treeNode(name)
            if name.is_dir():
                dirQueue.append(newNode)
                cRoot.add_child(newNode)
            elif name.is_file():
                cRoot.add_child(newNode)
```

```python
nArea_feeding(dirTree=dirTree)
```

```python
print(dirTree.name)
print(dirTree.path)
print(dirTree.is_dir)
print(dirTree.is_file)
if dirTree.is_file:
    print(dirTree.file_type)
print(dirTree.children)

a = dirTree.children.copy()
print(f"Lenth A: {len(a)}")
while a != []:
    b = a.pop(0)
    print(b.name)
    print(b.path)
    print(b.is_dir)
    print(b.is_file)
    if b.is_file:
        print(b.file_type)
    print(b.children)
```
